```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr
```

```python
load_dotenv(override=True)
# openai_api_key = os.getenv('OPENAI_API_KEY')
# openai = OpenAI()

gemini_api_key = os.getenv('GOOGLE_API_KEY')
gemini = OpenAI(api_key=gemini_api_key, base_url="https://generativelanguage.googleapis.com/v1beta/openai/")
```

```python
Model = 'gemini-2.5-flash-lite'  #based on client api provider select the available model
```

```python
system_message = "You are a helpful assistant whose name is cessna-AI. Answer as concisely as possible."
```

```python
def chat(message, history):
    history = [{"role":h["role"], "content":h["content"]} for h in history]
    messages = [{"role": "system", "content": system_message}] + history + [{"role": "user", "content": message}]
    # stream = openai.chat.completions.create(model=MODEL, messages=messages, stream=True)
    stream = gemini.chat.completions.create(model=Model, messages=messages, stream=True)
    response = ""
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        yield response
```

```python
gr.ChatInterface(fn=chat, type="messages", title="Cessna-AI").launch()
```
